```python
print("Total no.of ages:", vector1.size)
import matplotlib.pyplot as plt
x_axis = data['age'][(data['age'] >= 20) & (data['age'] <= 40)]
print("No.of ages which are between 20 to 40 is:",x_axis.size)
y_axis = data['smoker'][(data['age'] >= 20) & (data['age'] <= 40)]
plt.scatter(x_axis,y_axis)
plt.xlabel('age')
plt.ylabel('smoker')
plt.title('20 to 40 years age group smokers')
plt.show()
```
