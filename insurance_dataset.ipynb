```python
below_39_count = data['age'][(data['age'] < 39) & (data['smoker']==1)].count() # Added parentheses to call the count method
above_39_count = data['age'][(data['age'] > 39) & (data['smoker']==1)].count() # Added parentheses to call the count method
plt.bar(['Below 39', 'Above 39'], [below_39_count, above_39_count])
plt.title('below and above 39 age group smokers')
plt.ylabel('Number of Smokers')
plt.show()
```

```python
import matplotlib.pyplot as plt
smoker_expenses = data['expenses'][data['smoker'] == 1]
non_smoker_expenses = data['expenses'][data['smoker'] == 0]
plt.figure(figsize=(8, 6))
plt.hist(smoker_expenses, bins=30, alpha=0.5, label='Smokers')
plt.hist(non_smoker_expenses, bins=30, alpha=0.5, label='Non-Smokers')
plt.xlabel('Expenses')
plt.ylabel('Frequency')
plt.title('Distribution of Expenses for Smokers and Non-Smokers')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/insurance.csv')
smoker_by_gender = data.groupby('sex')['smoker'].value_counts().unstack()
smoker_by_gender.plot(kind='bar', figsize=(8, 6))
plt.title('Number of Smokers by Gender')
plt.xlabel('Gender (Encoded)')
plt.ylabel('Number of Smokers')
plt.xticks(rotation=0)
plt.legend(['Non-Smoker', 'Smoker'])
plt.show()
```
