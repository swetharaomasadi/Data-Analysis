```python
from scipy.stats import binom

# Parameters
n = 10  # Sample size
p = 0.3  # Probability of being a smoker
k = 4  # At least 4 smokers

# Compute P(X â‰¥ 4) using the complement rule
prob_at_least_4 = 1 - binom.cdf(k - 1, n, p)

print(f"Probability of at least 4 smokers: {prob_at_least_4:.4f}")
```

```python
# Parameters
n = 15  # Sample size
p = 0.2  # Probability of expenses > 20,000
k = 5  # Exactly 5 individuals

# Compute P(X = 5)
prob_exactly_5 = binom.pmf(k, n, p)

print(f"Probability of exactly 5 individuals having expenses > $20,000: {prob_exactly_5:.4f}")
```

```python
# Compute P(X â‰¤ 5)
prob_at_most_5 = binom.cdf(5, 15, 0.2)

print(f"Cumulative probability of at most 5 individuals: {prob_at_most_5:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 16)

# Compute CDF
cdf_values = binom.cdf(x, 15, 0.2)

# Plot
plt.step(x, cdf_values, where='mid', color='red', label="CDF")
plt.xlabel("Number of Individuals with Expenses > $20,000")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (n=15, p=0.2)")
plt.legend()
plt.grid()
plt.show()
```

```python
# Parameters
n = 50  # Sample size
p_southwest = 0.25
p_southeast = 0.30
p_northwest = 0.20
p_northeast = 0.35

# Simulate 1000 samples for each region
samples_southwest = binom.rvs(n, p_southwest, size=1000)
samples_southeast = binom.rvs(n, p_southeast, size=1000)
samples_northwest = binom.rvs(n, p_northwest, size=1000)
samples_northeast = binom.rvs(n, p_northeast, size=1000)

# Display first 5 samples for each region
print("Southwest Samples:", samples_southwest[:5])
print("Southeast Samples:", samples_southeast[:5])
print("Northwest Samples:", samples_northwest[:5])
print("Northeast Samples:", samples_northeast[:5])
```

```python
# Plot histogram for each region
plt.hist(samples_southwest, bins=20, alpha=0.6, label="Southwest", color="blue")
plt.hist(samples_southeast, bins=20, alpha=0.6, label="Southeast", color="green")
plt.hist(samples_northwest, bins=20, alpha=0.6, label="Northwest", color="red")
plt.hist(samples_northeast, bins=20, alpha=0.6, label="Northeast", color="purple")

plt.xlabel("Number of Smokers in Sample (n=50)")
plt.ylabel("Frequency")
plt.title("Distribution of Smokers per Region")
plt.legend()
plt.grid()
plt.show()
```

```python
# Expected number of smokers per region (E[X] = n * p)
expected_southwest = n * p_southwest
expected_southeast = n * p_southeast
expected_northwest = n * p_northwest
expected_northeast = n * p_northeast

# Compute mean from simulated data
mean_southwest = samples_southwest.mean()
mean_southeast = samples_southeast.mean()
mean_northwest = samples_northwest.mean()
mean_northeast = samples_northeast.mean()

# Print results
print(f"Expected vs Simulated Smokers:")
print(f"Southwest: Expected = {expected_southwest}, Simulated = {mean_southwest:.2f}")
print(f"Southeast: Expected = {expected_southeast}, Simulated = {mean_southeast:.2f}")
print(f"Northwest: Expected = {expected_northwest}, Simulated = {mean_northwest:.2f}")
print(f"Northeast: Expected = {expected_northeast}, Simulated = {mean_northeast:.2f}")
```

```python
# Define parameters
n_policyholders = 100  # Total number of policyholders
high_risk_threshold = binom.ppf(0.90, n_policyholders, 0.1)  # Top 10% threshold

print(f"Threshold number of high-risk policyholders: {high_risk_threshold}")
```

```python
# Parameters
n_sample = 30  # Sample size
p_high_risk = 0.1  # Probability of high-risk
k_high_risk = 8  # Exactly 8 high-risk individuals

# Compute P(X = 8)
prob_8_high_risk = binom.pmf(k_high_risk, n_sample, p_high_risk)

print(f"Probability of selecting exactly 8 high-risk individuals: {prob_8_high_risk:.6f}")
```

```python
# Values of X (number of high-risk individuals)
x = np.arange(0, 31)

# Compute PMF
pmf_values = binom.pmf(x, 30, 0.1)

# Plot PMF
plt.bar(x, pmf_values, color="orange", edgecolor="black")
plt.xlabel("Number of High-Risk Individuals")
plt.ylabel("Probability")
plt.title("Probability Mass Function (n=30, p=0.1)")
plt.xticks(x)
plt.grid()
plt.show()
```

```python
# Parameters
n_children = 12  # Sample size
p_more_than_3 = 0.4  # Probability of having more than 3 children

# Compute P(X â‰¥ 5) using complement rule: P(X â‰¥ 5) = 1 - P(X â‰¤ 4)
prob_at_least_5 = 1 - binom.cdf(4, n_children, p_more_than_3)

print(f"Probability of at least 5 individuals having more than 3 children: {prob_at_least_5:.6f}")
```

```python
# Simulate 1000 samples
simulated_samples = binom.rvs(12, 0.4, size=1000)

# Plot histogram
plt.hist(simulated_samples, bins=np.arange(0, 14) - 0.5, edgecolor="black", alpha=0.7)
plt.xlabel("Number of Individuals with More than 3 Children")
plt.ylabel("Frequency")
plt.title("Histogram of Simulated Binomial Distribution (n=12, p=0.4)")
plt.xticks(range(13))
plt.grid()
plt.show()
```

```python
# Compute theoretical mean and variance
theoretical_mean = 12 * 0.4  # E[X] = n * p
theoretical_variance = 12 * 0.4 * (1 - 0.4)  # Var(X) = n * p * (1 - p)

# Compute mean and variance from simulated data
simulated_mean = np.mean(simulated_samples)
simulated_variance = np.var(simulated_samples, ddof=1)  # Sample variance

# Print results
print(f"Theoretical Mean: {theoretical_mean:.2f}, Simulated Mean: {simulated_mean:.2f}")
print(f"Theoretical Variance: {theoretical_variance:.2f}, Simulated Variance: {simulated_variance:.2f}")
```
