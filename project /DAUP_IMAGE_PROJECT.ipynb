```python
import os

def print_dataset_summary_short(base_path, max_examples=2):
    print(f"📂 Scanning dataset at: {base_path}\n")

    total_normal = 0
    total_sick = 0

    for label in ["Normal", "Sick"]:
        label_path = os.path.join(base_path, label)
        image_count = 0
        folder_preview = []

        for root, dirs, files in os.walk(label_path):
            jpg_files = [f for f in files if f.lower().endswith('.jpg')]
            if jpg_files:
                image_count += len(jpg_files)
                if len(folder_preview) < max_examples:
                    relative_path = os.path.relpath(root, base_path)
                    folder_preview.append((relative_path, len(jpg_files)))

        print(f"📁 {label}: {image_count} images")
        for rel_path, count in folder_preview:
            print(f"   └── {rel_path}/ ({count} images)")

        print()

        if label == "Normal":
            total_normal = image_count
        else:
            total_sick = image_count

    print(f"✅ TOTAL IMAGES:\n   Normal: {total_normal}\n   Sick: {total_sick}")
    print(f"🧮 Total Combined: {total_normal + total_sick}")

# Set your dataset root
dataset_root = "/cad_mri_dataset"
print_dataset_summary_short(dataset_root)
```

```python
import os
import cv2
import numpy as np
import shutil
import random
from tqdm import tqdm

# Paths
original_root = "/cad_mri_dataset"
processed_root = "/processed_mri_dataset"
final_split_root = "/split_mri_dataset"

# Ensure reproducibility
random.seed(42)

# Step 1: Preprocess images and save to processed_mri_dataset
def preprocess_images():
    print("🔄 Starting image preprocessing...\n")
    image_count = 0
    failed_images = []

    for root, _, files in os.walk(original_root):
        for file in tqdm(files, desc=f"Processing in {os.path.basename(root)}"):
            if file.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                input_path = os.path.join(root, file)
                relative_path = os.path.relpath(input_path, original_root)
                output_path = os.path.join(processed_root, relative_path)

                try:
                    # Create output directories
                    os.makedirs(os.path.dirname(output_path), exist_ok=True)

                    # Load image
                    image = cv2.imread(input_path, cv2.IMREAD_GRAYSCALE)
                    if image is None:
                        failed_images.append(input_path)
                        continue

                    # Resize and normalize
                    image_resized = cv2.resize(image, (256, 256))
                    image_normalized = (image_resized / 255.0 * 255).astype(np.uint8)

                    # Save
                    cv2.imwrite(output_path, image_normalized)
                    image_count += 1

                except Exception as e:
                    failed_images.append(input_path)

    print(f"\n✅ Preprocessing complete: {image_count} images processed.")
    if failed_images:
        print(f"⚠️ {len(failed_images)} images failed to process and were skipped.")
        for path in failed_images[:5]:
            print("   -", path)

# Step 2: Split into train/valid/test
def split_dataset():
    print("\n📦 Starting dataset split (Deep Scan)...")
    categories = ['Normal', 'Sick']
    splits = {'train': 0.7, 'valid': 0.15, 'test': 0.15}

    for category in categories:
        all_images = []
        category_path = os.path.join(processed_root, category)

        # Deep scan
        for root, _, files in os.walk(category_path):
            for file in files:
                if file.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                    all_images.append(os.path.join(root, file))

        print(f"🔍 Found {len(all_images)} images in {category}")

        # Shuffle & split
        random.shuffle(all_images)
        train_count = int(len(all_images) * splits['train'])
        valid_count = int(len(all_images) * splits['valid'])

        split_data = {
            'train': all_images[:train_count],
            'valid': all_images[train_count:train_count + valid_count],
            'test': all_images[train_count + valid_count:]
        }

        # Copy to final folders
        for split in ['train', 'valid', 'test']:
            for img_path in tqdm(split_data[split], desc=f"Copying {category} -> {split}"):
                filename = os.path.basename(img_path)
                dest_dir = os.path.join(final_split_root, split, category)
                os.makedirs(dest_dir, exist_ok=True)

                # Ensure unique filenames if needed
                dest_path = os.path.join(dest_dir, filename)
                base, ext = os.path.splitext(filename)
                counter = 1
                while os.path.exists(dest_path):
                    dest_path = os.path.join(dest_dir, f"{base}_{counter}{ext}")
                    counter += 1

                shutil.copy(img_path, dest_path)

    print("✅ Deep dataset split completed.")

# Step 3: Count images in final splits
def count_images_in_split():
    print("\n📊 Image Summary:\n")
    total = 0
    for split in ['train', 'valid', 'test']:
        print(f"📁 {split.upper()}:")
        for category in ['Normal', 'Sick']:
            folder = os.path.join(final_split_root, split, category)
            count = sum(len(files) for _, _, files in os.walk(folder))
            print(f"   └── {category}: {count} images")
            total += count
    print(f"\n🧮 Total Combined Images in Split: {total}\n")

# Run all steps
if __name__ == "__main__":
    preprocess_images()
    split_dataset()
    count_images_in_split()
```

```python
import os
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Paths to the split folders
data_dir = "/split_mri_dataset"
batch_size = 32  # You can change as needed

# Define transforms for preprocessing
transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=1),  # Already grayscale, but ensures consistency
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])   # Normalize to [-1, 1]
])

# Load datasets
train_dataset = datasets.ImageFolder(os.path.join(data_dir, 'train'), transform=transform)
valid_dataset = datasets.ImageFolder(os.path.join(data_dir, 'valid'), transform=transform)
test_dataset  = datasets.ImageFolder(os.path.join(data_dir, 'test'),  transform=transform)

# Create DataLoaders
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False)
test_loader  = DataLoader(test_dataset,  batch_size=batch_size, shuffle=False)

# Class mapping
print("📂 Class Index to Label:", train_dataset.classes)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class MRICNN(nn.Module):
    def __init__(self):
        super(MRICNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 16, kernel_size=3, padding=1)  # 1 input channel (grayscale)
        self.pool1 = nn.MaxPool2d(2, 2)

        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, padding=1)
        self.pool2 = nn.MaxPool2d(2, 2)

        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.pool3 = nn.MaxPool2d(2, 2)

        self.fc1 = nn.Linear(64 * 32 * 32, 128)  # After 3 poolings on 256x256 image
        self.dropout = nn.Dropout(0.4)
        self.fc2 = nn.Linear(128, 2)  # 2 classes: Normal and Sick

    def forward(self, x):
        x = self.pool1(F.relu(self.conv1(x)))  # [1, 256, 256] → [16, 128, 128]
        x = self.pool2(F.relu(self.conv2(x)))  # → [32, 64, 64]
        x = self.pool3(F.relu(self.conv3(x)))  # → [64, 32, 32]
        x = x.view(-1, 64 * 32 * 32)           # Flatten
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x

# Instantiate model
model = MRICNN()
print("✅ CNN model created.")
```

```python
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader

# Define image transformations
transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=1),
    transforms.Resize((256, 256)),
    transforms.ToTensor(),  # Converts [0,255] → [0,1]
    transforms.Normalize((0.5,), (0.5,))  # Mean and Std for grayscale
])

# Paths to your split data
split_root = "/split_mri_dataset"

# Load datasets
train_dataset = ImageFolder(root=f"{split_root}/train", transform=transform)
valid_dataset = ImageFolder(root=f"{split_root}/valid", transform=transform)
test_dataset  = ImageFolder(root=f"{split_root}/test",  transform=transform)

# DataLoaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=32, shuffle=False)
test_loader  = DataLoader(test_dataset, batch_size=32, shuffle=False)

print(f"✅ DataLoaders ready.")
print(f"🧪 Train: {len(train_dataset)} | Valid: {len(valid_dataset)} | Test: {len(test_dataset)}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Simple CNN for binary classification
class MRIClassifier(nn.Module):
    def __init__(self):
        super(MRIClassifier, self).__init__()
        self.model = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),  # 128x128

            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),  # 64x64

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),  # 32x32

            nn.Flatten(),
            nn.Linear(64 * 32 * 32, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, 2)  # 2 classes: Normal, Sick
        )

    def forward(self, x):
        return self.model(x)

# Instantiate model, loss, and optimizer
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = MRIClassifier().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Tracking lists
train_losses = []
valid_losses = []
train_accuracies = []
valid_accuracies = []

# Training loop
def train_model(epochs=10):
    for epoch in range(epochs):
        model.train()
        train_loss = 0
        correct = 0
        total = 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()

            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            train_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        train_losses.append(train_loss / len(train_loader))
        train_accuracies.append(100 * correct / total)

        # Validation
        model.eval()
        val_loss = 0
        val_correct = 0
        val_total = 0
        with torch.no_grad():
            for images, labels in valid_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()

        valid_losses.append(val_loss / len(valid_loader))
        valid_accuracies.append(100 * val_correct / val_total)

        print(f"📅 Epoch {epoch+1}/{epochs} | "
              f"Train Loss: {train_losses[-1]:.4f}, Acc: {train_accuracies[-1]:.2f}% | "
              f"Valid Loss: {valid_losses[-1]:.4f}, Acc: {valid_accuracies[-1]:.2f}%")

train_model(epochs=10)
```

```python
def evaluate_model(model, dataloader, criterion, device, mode="Validation"):
    model.eval()  # Set model to evaluation mode
    total_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():  # No gradient needed
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            total_loss += loss.item() * images.size(0)
            _, predicted = torch.max(outputs, 1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)

    avg_loss = total_loss / total
    accuracy = correct / total * 100
    print(f"📊 {mode} Loss: {avg_loss:.4f}, Accuracy: {accuracy:.2f}%")
    return avg_loss, accuracy

# Evaluate after training
print("\n🔍 Final Evaluation:")
evaluate_model(model, valid_loader, criterion, device, mode="Validation")
evaluate_model(model, test_loader, criterion, device, mode="Test")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def plot_confusion_matrix(model, dataloader, device, class_names):
    model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    cm = confusion_matrix(all_labels, all_preds)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
    disp.plot(cmap=plt.cm.Blues)
    plt.title("🧠 Confusion Matrix")
    plt.show()

# Call it with class names
plot_confusion_matrix(model, test_loader, device, class_names=["Normal", "Sick"])
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np

def plot_roc_curve(model, dataloader, device):
    model.eval()
    y_true = []
    y_scores = []

    with torch.no_grad():
        for images, labels in dataloader:
            images = images.to(device)
            outputs = model(images)
            probs = torch.softmax(outputs, dim=1)[:, 1]  # Probability for class 'Sick'

            y_scores.extend(probs.cpu().numpy())
            y_true.extend(labels.cpu().numpy())

    # ROC curve
    fpr, tpr, _ = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic - ROC Curve')
    plt.legend(loc='lower right')
    plt.grid()
    plt.show()

# Call it for test set
plot_roc_curve(model, test_loader, device)
```

```python
from scipy.stats import norm
import numpy as np

# Convert to numpy arrays
train_acc = np.array(train_accuracies)
valid_acc = np.array(valid_accuracies)

# Means and standard errors
mean1, mean2 = train_acc.mean(), valid_acc.mean()
std1, std2 = train_acc.std(ddof=1), valid_acc.std(ddof=1)
n1, n2 = len(train_acc), len(valid_acc)

# Z-Test statistic
z = (mean1 - mean2) / np.sqrt((std1**2 / n1) + (std2**2 / n2))
p_value_z = 2 * (1 - norm.cdf(abs(z)))

print("\n🧪 Z-Test Results:")
print(f"Z = {z:.4f}, p-value = {p_value_z:.4f}")
if p_value_z < 0.05:
    print("✅ Significant difference between training and validation accuracies.")
else:
    print("⚖️ No significant difference between training and validation accuracies.")
```

```python
from scipy.stats import ttest_ind

t_stat, p_value_t = ttest_ind(train_acc, valid_acc, equal_var=False)

print("\n🧪 T-Test Results:")
print(f"T = {t_stat:.4f}, p-value = {p_value_t:.4f}")
if p_value_t < 0.05:
    print("✅ Significant difference between training and validation accuracies.")
else:
    print("⚖️ No significant difference between training and validation accuracies.")
```
